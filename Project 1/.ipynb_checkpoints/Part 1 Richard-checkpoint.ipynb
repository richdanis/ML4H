```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
train_df = pd.read_csv('heart_failure/train_val_split.csv')  
y_train = train_df.loc[:,'HeartDisease']
train_df = train_df.drop(columns=['HeartDisease'])
```

### Columns

```python
train_df.columns
```

### Checking for Nans

```python
train_df.isnull().sum()
```

### Converting the string features to numerical features

There are 5 feature columns with string values. <br>
These need to be converted to numerical values in order to use them.

Sex: M,F <br>
ChestPainType: 'ATA', 'ASY', 'NAP', 'TA' <br>
RestingECG: 'Normal', 'ST', 'LVH' <br>
ExerciseAngina: 'N', 'Y' <br>
ST_Slope: 'Up', 'Flat', 'Down'

```python
train_df = train_df.replace({'M': 0.0, 'F': 1.0, 'N': 0.0, 'Y': 1.0})
```

```python
ata = (train_df.loc[:,'ChestPainType'] == 'ATA').to_numpy().astype(float)
asy = (train_df.loc[:,'ChestPainType'] == 'ASY').to_numpy().astype(float)
nap = (train_df.loc[:,'ChestPainType'] == 'NAP').to_numpy().astype(float)
ta = (train_df.loc[:,'ChestPainType'] == 'TA').to_numpy().astype(float)

train_df['ATA'] = ata
train_df['ASY'] = asy
train_df['NAP'] = nap
train_df['TA'] = ta
train_df = train_df.drop(columns=['ChestPainType'])
```

```python
normal = (train_df.loc[:,'RestingECG'] == 'Normal').to_numpy().astype(float)
st = (train_df.loc[:,'RestingECG'] == 'ST').to_numpy().astype(float)
lvh = (train_df.loc[:,'RestingECG'] == 'LVH').to_numpy().astype(float)

train_df['Normal'] = normal
train_df['ST'] = st
train_df['LVH'] = lvh
train_df = train_df.drop(columns=['RestingECG'])
```

```python
up = (train_df.loc[:,'ST_Slope'] == 'Up').to_numpy().astype(float)
flat = (train_df.loc[:,'ST_Slope'] == 'Flat').to_numpy().astype(float)
down = (train_df.loc[:,'ST_Slope'] == 'Down').to_numpy().astype(float)

train_df['Up'] = up
train_df['Flat'] = flat
train_df['Down'] = down
train_df = train_df.drop(columns=['ST_Slope'])
```

```python
train_df['HeartDisease'] = y_train
```

### Histograms

```python
plt.rcParams["figure.figsize"] = (20,12)
hist = train_df.hist()
plt.tight_layout()
```

### Fixing Cholesterol Distribution

```python
chol_mean = train_df['Cholesterol'].replace(0, np.NaN).mean()
chol_std = train_df['Cholesterol'].replace(0, np.NaN).std()
train_df['Cholesterol'] = train_df['Cholesterol'].replace(0, chol_mean)
```

### Normalizing

Only normalize Age, RestingBP, Cholesterol and MaxHR (and maybe Oldpeak)

```python
names = ['Age', 'RestingBP', 'Cholesterol', 'MaxHR', 'Oldpeak']
for name in names:
    if name == 'Cholesterol':
        train_df[name] = (train_df[name] - train_df[name].mean()) / chol_std
    else:
        train_df[name] = (train_df[name] - train_df[name].mean()) / train_df[name].std()
```

### Correlation Matrix

```python
mask = np.tril(np.ones_like(train_df.corr(), dtype=bool))
cmap = sns.diverging_palette(230, 20, as_cmap=True)
sns.heatmap(train_df.corr(), mask=mask, cmap=cmap)
```
